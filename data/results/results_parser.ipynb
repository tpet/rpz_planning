```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
xls_files = [file for file in os.listdir() if 'xls' in file]
```

```python
data = pd.read_excel('./simple_cave_02.xls')
data_trajopt = pd.read_excel('./simple_cave_02_trajopt.xls')
```

```python
data.head()
```

```python
metrics = [key for key in data if not 'Time stamp' in key]

plt.figure(figsize=(10, 7*len(metrics)))
for i, metric in enumerate(metrics):
    plt.subplot(len(metrics), 1, i+1)
    plt.title(metric)
    plt.plot(data['Time stamp'], data[metric], label='naex')
    plt.plot(data_trajopt['Time stamp'], data_trajopt[metric], label='naex_trajopt')
    plt.grid()
    plt.xlabel('time, [sec]')
    plt.legend();
```
