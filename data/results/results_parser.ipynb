```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
xls_files = [file for file in os.listdir() if 'xls' in file]
```

```python
data = pd.read_excel('./simple_cave_02.xls')
data_trajopt = pd.read_excel('./simple_cave_02_trajopt.xls')
```

```python
data.head()
```

```python
for metric in data:
    if metric == 'Time stamp':
        continue
    plt.figure()
    plt.title(metric)
    plt.plot(data['Time stamp'], data[metric], label='naex')
    plt.plot(data_trajopt['Time stamp'], data_trajopt[metric], label='naex_trajopt')
    plt.grid()
    plt.xlabel('time, [sec]')
    plt.legend();
```
